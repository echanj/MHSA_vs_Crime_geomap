```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns

# Cause plots to be displayed in the notebook:
%pylab inline
%matplotlib inline

from google.cloud import bigquery
```

```python
import sqlite3
from pandas.io import sql
sqlite_db = 'miniproject_db.sqlite'
conn = sqlite3.connect(sqlite_db) 
c = conn.cursor()
```

```python
key_path = './marine-fusion-381400-bcd419f0d93b.json'
```

```python
myclient = bigquery.Client.from_service_account_json(key_path, project = 'marine-fusion-381400')
print(myclient.project)
mydatasets = list(myclient.list_datasets())
mydatasets
```

```python
sql = '''SELECT COUNT(1) as num_rows  FROM `bigquery-public-data.samples.shakespeare` '''
query_job = myclient.query(sql)
```

```python
sql = ''' SELECT  word, word_count, corpus  FROM `bigquery-public-data.samples.shakespeare` LIMIT 10; '''
query_job = myclient.query(sql)
```

```python
for row in query_job:
    assert row[0] == row['word']  
    assert row[1] == row['word_count']  
    assert row[2] == row['corpus']  
    print(row[0],row[1],row[2])
```

```python
df = query_job.to_dataframe()
print(df)    
```

```python
sql = ''' SELECT Address,City,State_Abbr,County_Name,Postal_Code,
Degree_of_Shortage,Status_Description,Designation_Pop,
Estimated_Underserved_Pop,HPSA_Score, Designation_Date,
Component_Designation_Date,HPSA_Withdrawn_Date,
Geocoding_Primary_Y as latitude,
Geocoding_Primary_X as longitude   FROM `bigquery-public-data.sdoh_hrsa_shortage_areas.hpsa_mental_health` 
WHERE 
 NOT Address IS NULL
 AND 
 NOT Degree_of_Shortage IS NULL
 AND 
 NOT HPSA_Score IS NULL
 AND 
 NOT Designation_Pop IS NULL
 -- AND
 -- NOT Status_Description = 'Withdrawn'
LIMIT 10; '''
query_job = myclient.query(sql)
```

```python
df = query_job.to_dataframe()
df
```

```python
df.to_sql('hpsa_mental_health_short', con = conn, if_exists = 'replace', index = False)
```
